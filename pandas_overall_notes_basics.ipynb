# label based indexing--loc

# ranking the data frames


```python
import pandas as pd
df=pd.Series([11,3443,142,45,44,5],index=["fri","tue","wed","sun","sat","mon"])
df.rank(ascending=False).sort_values(ascending=False)
df.rank(ascending=False).sort_values()
```
